```python
input_dir = "data-internet/ABUZZ_preprocessed_database/____not_mosquito/"  # Change this to the input folder
output_dir = "not_mosquito/ABUZZ/"  # Change this to the output folder
n=20000 # number of files to copy
```

```python
import os
import glob
import random
import shutil

def copy_random_wavs(input_dir, output_dir, n=10000, seed=42):
    """
    Recursively lists all .wav files in a folder and subfolders, shuffles them randomly, 
    and copies the first N files to the output folder.

    Args:
        input_dir (str): The path to the input folder.
        output_dir (str): The path to the output folder.
        n (int): The number of files to copy.
        seed (int): The seed value for randomness for reproducibility.
    """
    # Lists all .wav files recursively
    wav_files = glob.glob(os.path.join(input_dir, "**", "*.wav"), recursive=True)

    # Set random order based on seed
    random.seed(seed)
    random.shuffle(wav_files)

    # Select the first N files
    selected_files = wav_files[:n]

    # Ensure the output folder exists
    os.makedirs(output_dir, exist_ok=True)

    # Copy files to the output folder
    for file in selected_files:
        base_name = os.path.basename(file)  # Just the filename
        dest_path = os.path.join(output_dir, base_name)
        shutil.copy(file, dest_path)

    print(f"Copied {len(selected_files)} .wav files to the {output_dir} folder.")



```

```python
copy_random_wavs(input_dir, output_dir, n=n, seed=42)
```
